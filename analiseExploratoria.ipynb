```python
# Integrantes:
# Gustavo - 10418552
# Otto - 10418170
# Renan - 10417093
# Rodrigo - 10417090

import pandas as pd
import matplotlib.pyplot as plt

# Carregar dataset
df = pd.read_csv("dados_juridicos.csv", encoding="utf-8", engine="python")

print("Dataset carregado:")
print(df.head())

# --------------------------
# Análise 1: tamanho dos textos
# --------------------------
df["num_palavras"] = df["texto_original"].apply(lambda x: len(x.split()))
df["num_caracteres"] = df["texto_original"].apply(len)

print("\nMédia de palavras:", df["num_palavras"].mean())
print("Média de caracteres:", df["num_caracteres"].mean())

# --------------------------
# Análise 2: distribuição por tipo
# --------------------------
print("\nDistribuição por tipo:")
print(df["tipo_texto"].value_counts())

# --------------------------
# Gráfico
# --------------------------
df["tipo_texto"].value_counts().plot(kind="bar")
plt.title("Distribuição dos tipos de texto")
plt.xlabel("Tipo")
plt.ylabel("Quantidade")
plt.show()
```
